```python
# Write a Pandas program to compare the elements of two Pandas series. 
# Create the two series
x = pd.Series([4, 65, 436, 3, 9])
y = pd.Series([7, 0, 3, 897, 9])
# Compare the elements of the two series
comparison = x == y
# Print the comparison result
print(comparison)
```

```python
# Write a Pandas program to add, subtract, multiply and divide two Pandas series. 
# Create the two series
a = pd.Series([2, 4, 6, 8, 14])
b = pd.Series([1, 3, 5, 7, 9])
# Addition
addition = a + b
print("Addition:")
print(addition)
# Subtraction
subtraction = a - b
print("\nSubtraction:")
print(subtraction)
# Multiplication
multiplication = a * b
print("\nMultiplication:")
print(multiplication)
# Division
division = a / b
print("\nDivision:")
print(division)
```

```python
# Write a Pandas program to convert a given series to an array.
# Sample series
g = pd.Series(['Love', 800, 'Joy', 789.9, 'Peace', True])
# Convert the series to an array
h = np.array(g)
# Print the resulting array
print(h)
```

```python
# Write a Pandas program to display the most frequent value in the given series 
# and replace everything else as ‘Other’ in the series.
df=pd.read_csv("AfricaCupofNationsMatches.csv")
# Calculate the most frequent value
most_frequent_value = df['HomeTeamGoals'].mode()[0]
# Replace everything else as 'Other'
df['HomeTeamGoals'] = df['HomeTeamGoals'].where(df['HomeTeamGoals'] == most_frequent_value, 'Other')
# Print the modified series
print(df['HomeTeamGoals'])
```

```python
# Write a Pandas program to read the given csv file
df=pd.read_csv("AfricaCupofNationsMatches.csv")
df
```

```python
# Write a Pandas program to get the first 7 rows of your data frame.
first_7_rows = df.head(7)
# Print the first 7 rows
print(first_7_rows)
```

```python
# Write a Pandas program to select the ‘HomeTeam’, ‘AwayTeam’, ‘HomeTeamGoals’ 
#and ‘AwayTeamGoals’ columns from your data frame.
# Select the desired columns
selected_columns = df[['HomeTeam', 'AwayTeam', 'HomeTeamGoals', 'AwayTeamGoals']]
# Print the selected columns
print(selected_columns)
```

```python
# Write a Pandas program to select the rows where Egypt appears
# Select rows where 'Egypt' appears
selected_rows = df[df.eq('Egypt').any(axis=1)]
# Print the selected rows
print(selected_rows)
```

```python
# Write a Pandas program to count the number of rows and columns of your data frame.
# Count the number of rows and columns
num_rows = df.shape[0]
num_columns = df.shape[1]
# Print the number of rows and columns
print("Number of rows:", num_rows)
print("Number of columns:", num_columns)
```

```python
# Write a Pandas program to select the rows where the ‘Attendance’ is missing.
# Select rows where 'Attendance' is missing
select = df[df['Attendance'].isna()]
# Print the selected rows
print(select)
```

```python
# Write a Pandas program to select the rows where the ‘HomeTeamGoals’ are between 3 and 6 inclusive.
# Select rows where 'HomeTeamGoals' is between 3 and 6 (inclusive)
ng = df[(df['HomeTeamGoals'] >= 3) & (df['HomeTeamGoals'] <= 6)]
# Print the selected rows
print(ng)
```

```python
# Write a Pandas program to change the “AwayTeamGoals” in the 3rd row to 10.
df.at[2, "AwayTeamGoals"] = 10
# Print the updated DataFrame
print(df.head(3))
```

```python
# Write a Pandas program to sort the DataFrame first by ‘HomeTeam’ in
# ascending order, then by ‘HomeTeamScores’ in descending order.
df1 = df.sort_values(by=['HomeTeam', 'HomeTeamGoals'], ascending=[True, False])
# Print the sorted DataFrame
print(df1)
```
